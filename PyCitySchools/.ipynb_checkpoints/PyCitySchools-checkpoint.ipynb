```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])


school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Count the onumber of schools in the data set
school_count = len(school_data_complete['school_name'].unique())
#create a dataframe for the information 
Schools_df = pd.DataFrame({"Total Schools": [school_count]})
#count total number of students
total_students = school_data_complete['student_name'].count()
#add total students to dataframe
Schools_df['Total Students'] = total_students
#calculate total budget
total_budget = sum(school_data_complete['budget'].unique())
#add total budget for all schools to dataframe
Schools_df['Total Budget'] = total_budget
#Calc avg math score for all students
avg_math = school_data_complete['math_score'].mean()
#add avg math score to dataframe
Schools_df['Average Math Score']=round(avg_math,2)
#avgerage reading grade
avg_reading = school_data_complete['reading_score'].mean()
#add avg reading to dataframe
Schools_df['Average Reading Score']=round(avg_reading,2)
#students who pass math into True series
studentspassingmath = school_data_complete.apply(lambda x: True if x['math_score'] >= 70 else False, axis=1)
#percent students passing math number of True in series
passingmathpercent = round((((len(studentspassingmath[studentspassingmath == True].index))/total_students)*100),2)
#add above to dataframe
Schools_df['% Passing Math'] = passingmathpercent.astype(str) + '%'
#students who pass reading
studentspassingreading = school_data_complete.apply(lambda x: True if x['reading_score'] >= 70 else False, axis = 1)
#percent students passing reading
passingreadingpercent = round(((len(studentspassingreading[studentspassingreading == True].index)/total_students)*100),2)
#add reading passing percet to dataframe
Schools_df['% Passing Reading'] = passingreadingpercent.astype(str) + '%'
# students passing both
passingboth = school_data_complete.apply(lambda x: True if x['reading_score'] >= 70 and x['math_score'] >= 70 else False, axis = 1)
#calc passing both percent
passingbothpercent = round(((len(passingboth[passingboth == True].index)/total_students)*100),2)
#adding to dataframe
Schools_df['% Overall Passing'] = passingbothpercent.astype(str) + "%"

Schools_df
```
