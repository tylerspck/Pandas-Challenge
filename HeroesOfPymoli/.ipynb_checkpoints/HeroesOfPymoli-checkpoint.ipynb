### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

### Note
* Instructions have been included for each segment. You do not have to follow them exactly, but they are included to help you think through the steps.

```python
# Dependencies and Setup
import pandas as pd
pd.options.display.float_format = '${:,.2f}'.format

# File to Load (Remember to Change These)
csvfile = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(csvfile)

purchase_data_df.head(10)
```

## Player Count

```python
playercount = len(purchase_data_df["SN"].unique())

playercount_df = pd.DataFrame({"Total Players": [playercount]})

playercount_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
NumberofItems = len(purchase_data_df["Item Name"].unique())
Averageprice = sum(purchase_data_df['Price'].unique())/len(purchase_data_df['Price'].unique())
totalpurchases = len(purchase_data_df['Item Name'])
totalprofit = sum(purchase_data_df["Price"])

cNumberofItems = round(NumberofItems,2)
cAverageprice = round(Averageprice, 2)
ctotalpurchases = round(totalpurchases, 2)
ctotalprofit = round(totalprofit, 2)

Purchase_Analysis_df = pd.DataFrame({"Total Unique Items": [cNumberofItems],
                                    "Average Price": [cAverageprice],
                                    "Number of Purchases": [ctotalpurchases],
                                    "Total Revenue": [ctotalprofit]
                                    }
                                   )

Purchase_Analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
users_df = pd.DataFrame(purchase_data_df.groupby("Gender")["SN"].nunique())
users_df.columns = ['Total Users']
percent = users_df['Total Users']/playercount*100
percent2 = round(percent,2)
users_df['Percent of Users'] = percent2.astype(str) + '%'
users_df = users_df.reset_index(drop = False)
# users_df = user_df.set_index('')

users_df
```

## Puchasing Analysis (Gender)



* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
Number_Purchases_per_Gender = purchase_data_df.groupby("Gender")['Price'].count()

Purchashing_Gender_Df= pd.DataFrame(Number_Purchases_per_Gender)

Purchashing_Gender_Df = Purchashing_Gender_Df.rename(columns={'Price':'Purchase Count'})

average_purchase = purchase_data_df.groupby('Gender')['Price'].mean()

Purchashing_Gender_Df['Average Purchase Price'] = average_purchase

total_purchase = purchase_data_df.groupby('Gender')['Price'].sum()

Purchashing_Gender_Df['Total Purchase Value'] = total_purchase

users_by_gender = purchase_data_df.groupby("Gender")["SN"].nunique()

total_purchase_per_user = Purchashing_Gender_Df['Total Purchase Value']/users_by_gender

Purchashing_Gender_Df['Avg Total Purchase per Person'] = total_purchase_per_user

Purchashing_Gender_Df = Purchashing_Gender_Df.reset_index(drop = False)
Purchashing_Gender_Df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 40000]

age_groups =['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

purchase_data_df['Age Range'] = pd.cut(purchase_data_df['Age'], bins, labels=age_groups)

player_age_range = purchase_data_df.groupby("Age Range")['SN'].nunique()

Age_Dem_df = pd.DataFrame(player_age_range)

Age_Dem_df = Age_Dem_df.rename(columns={'SN':'Total Count'})

Percent_Age = round((Age_Dem_df['Total Count']/playercount)*100,2)

Age_Dem_df['Percentage of Players'] = Percent_Age.astype(str) + '%'

Age_Dem_df = Age_Dem_df.reset_index(drop = False)

Age_Dem_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_count_age = purchase_data_df.groupby('Age Range')['Price'].count()

Purchase_Age_df=pd.DataFrame(purchase_count_age)

Purchase_Age_df=Purchase_Age_df.rename(columns={'Price':'Purchase Count'})

avg_price_age = purchase_data_df.groupby('Age Range')['Price'].mean()

Purchase_Age_df['Average Purchase Price'] = avg_price_age

total_spent_age = purchase_data_df.groupby('Age Range')['Price'].sum()

Purchase_Age_df['Total Purchase Value']=total_spent_age

avg_spent_per_person_age = Purchase_Age_df['Total Purchase Value']/player_age_range

Purchase_Age_df['Avg Total Purchase per Person']=avg_spent_per_person_age

Purchase_Age_df=Purchase_Age_df.reset_index(drop = False)

Purchase_Age_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
SN_purchase_count = purchase_data_df.groupby('SN')['Price'].count()
Top_Spenders_df=pd.DataFrame(SN_purchase_count)
Top_Spenders_df=Top_Spenders_df.rename(columns={'Price':'Purchase Count'})
SN_purchase_sum = purchase_data_df.groupby('SN')['Price'].sum()
avg_purchase_price = SN_purchase_sum/SN_purchase_count
Top_Spenders_df['Average Purchase Price'] = avg_purchase_price
Top_Spenders_df['Total Purchase Value']=SN_purchase_sum


Top_Spenders_sorted_df=Top_Spenders_df.sort_values('Total Purchase Value', ascending = False)
Top_Spenders_sorted_df=Top_Spenders_sorted_df.reset_index(drop = False)


Top_Spenders_sorted_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Item_purchased = purchase_data_df.groupby('Item ID')['Item Name'].unique()
Most_Pop_Items_df = pd.DataFrame(Item_purchased)
Item_purchased_count = purchase_data_df.groupby('Item ID')['Price'].count()
Most_Pop_Items_df['Purchase Count']=Item_purchased_count
Item_price = purchase_data_df.groupby('Item ID')['Price'].mean()
Most_Pop_Items_df['Price']=Item_price
Item_price_total = purchase_data_df.groupby('Item ID')['Price'].sum()
Most_Pop_Items_df['Total Purchase Value'] = Item_price_total




# Item_name=purchase_data_df.groupby('Item ID')['Item Name']
# Most_Pop_Items_df['Item Name']=Item_name

Most_Popsort_df = Most_Pop_Items_df.sort_values('Purchase Count', ascending = False)
Most_Popsort_df=Most_Popsort_df.reset_index(drop = False)



Most_Popsort_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
Most_Profitable_Items_df = Most_Pop_Items_df.sort_values('Total Purchase Value', ascending = False)
Most_Profitable_Items_df = Most_Profitable_Items_df.reset_index(drop = False)
Most_Profitable_Items_df.head()
```
